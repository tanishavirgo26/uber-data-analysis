```python
data['Date/Time'] = pd.to_datetime(data['Date/Time'], format="%m/%d/%Y %H:%M:%S")
data['DayOfWeekNum'] = data['Date/Time'].dt.dayofweek

data['MonthDayNum'] = data['Date/Time'].dt.day
data['HourOfDay'] = data['Date/Time'].dt.hour
```
